```python
# Berikut ini adalah library yang dibutuhkan untuk proses scraping data review google playstore:
import numpy as np
import pandas as pd
from google_play_scraper import app
from google_play_scraper import Sort, reviews
from google_play_scraper import permissions
```

```python
# Proses Scraping aplikasi Livin by mandiri
scrap_review, token = reviews(
    'id.bmri.livin',  # ID aplikasi (didapatkan melalui link aplikasi di google)
    lang='id',               
    country='id',              
    sort=Sort.MOST_RELEVANT,   
    count=100, # ambil review sebanyak 100
    filter_score_with=None        
)
```

```python
# Menampilkan data yang berhasil diambil 
reviewslivin_df = pd.DataFrame(scrap_review)
reviewslivin_df.head()
```

```python
# Banyak data yang berhasil diambil
len(reviewslivin_df)
```

```python
# Menampilkan review yang rentangnya dari tanggal 23 Januari 2026 hingga 26 Februari 2026
review2 = reviewslivin_df[(reviewslivin_df['at']>"2026-01-23") & (reviewslivin_df['at']<"2026-02-26")]

review2
```

```python
# Mengubah hasil ke DataFrame
apps_df = pd.DataFrame(review2)

# Menyimpan ke CSV
apps_df.to_csv("hasil_Scraping_Livin(Jan-Feb26).csv", index=False)
```
